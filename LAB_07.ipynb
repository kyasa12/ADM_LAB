# CLUSTERING

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage



```

```python
file_path = "/content/student dataset.xlsx"
df = pd.read_excel(file_path, sheet_name="Sheet1")
```

```python
features = ["GPA", "Study Hours/Week", "Match Score"]
df_numeric = df[features]
```

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_numeric)
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
df["Cluster"] = kmeans.fit_predict(df_scaled)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df["Study Hours/Week"], y=df["Match Score"], hue=df["Cluster"], palette="viridis", s=100)
plt.xlabel("Study Hours/Week")
plt.ylabel("Match Score")
plt.title("K-Means Clustering of Students")
plt.legend(title="Cluster")
plt.show()
```

```python
linked = linkage(df_scaled, method='ward')
```

```python
plt.figure(figsize=(12, 6))
dendrogram(linked, labels=df.index, leaf_rotation=90, leaf_font_size=10)
plt.title("Hierarchical Clustering Dendrogram")
plt.xlabel("Student Index")
plt.ylabel("Distance")
plt.show()
```
